```python
# PYTHON NAMING CONVENTIONS
uma_123 = 50
print(uma_123)

# $uma12 = 5
# print($uma12) --------------> SINGLE LINE COMMENT

"""special characters are not allowed in identifiers
we will get a syntax error                      ------------> MULTIPLE LINE COMMENTS
"""

# 123uma = 56
# print(123uma)
"""identifiers cannot begin with digits(numbers)
# we will get  syntax error : invalid decimal literal """

# uma = 1289
# print(UMA)
"""identifiers are case sensitive we shouldn't change lower to upper or upper to lower even a single letter
it will show NameError: name 'UMA' is not defined """

# for = 30
# print(for)
""" we should not use keywords as identifiers
we will get SyntaxError: invalid syntax"""

# student id = 45
# print(student id)
"""we should not leave spaces in between identifier
we get SyntaxError: invalid syntax"""
```

```python
# PYTHON VARIABLES
# Creating a single variable and print it

name = "uma"
print(name)
print()

# creating two variables

name = 'uma'
age = 20
print(name)
print(age)
print()

# creating 2 variables and using one print statement

name = 'uma'
age = 20
print(name, age)
print()

# creating a variable with meaningful text

name = 'uma'
print('I am', name)
```

```python
# creating multiple values in single line

a, b, c = 10, 20, 30
print(a)
print(b)
print(c)
print(a,b,c)

# a, b, c = 10, 20
# print(a,b,c)
""" number of variables should be equal to number of values otherwise
we will get ValueError: not enough values to unpack (expected 3, got 2)"""

```

```python
# assigning single value to multiple variables at a time

a = b = c = 20
print(a)
print(b)
print(c)
print(a,b,c)
```

```python
# value re-initialization
a = 4
print("initial value")
print(a)

a = "uma"
print('after re-initialization')
print(a)
```
